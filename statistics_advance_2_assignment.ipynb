1)What is hypothesis testing in statistics
 - Hypothesis testing is a form of statistical inference that uses data from a sample to draw conclusions about a population parameter or a population probability distribution. First, a tentative assumption is made about the parameter or distribution. This assumption is called the null hypothesis and is denoted by H0

2) What is the null hypothesis, and how does it differ from the alternative hypothesis?
  - The null hypothesis and alternative hypothesis are two statements used in statistical testing to explore a research question. The null hypothesis (H0) proposes that there is no significant effect or relationship between variables being studied, while the alternative hypothesis (H1 or Ha) suggests that a significant effect or relationship does exist. The goal of hypothesis testing is to determine whether there is enough evidence to reject the null hypothesis in favor of the alternative hypothesis.


3)What is the significance level in hypothesis testing, and why is it important?
  - In hypothesis testing, the significance level (alpha, denoted as Î±) is the probability of incorrectly rejecting the null hypothesis when it's actually true. It's a threshold used to determine statistical significance and is crucial because it sets the maximum acceptable risk of making a false positive conclusion, or Type I error.   


4)What does a P-value represent in hypothesis testing?
  - In hypothesis testing, a p-value represents the probability of observing data as extreme as, or more extreme than, the data actually observed, assuming the null hypothesis is true. Essentially, it quantifies how likely the observed results are due to random chance if there were no real effect.   


5)How do you interpret the P-value in hypothesis testing?
 - The P-value in hypothesis testing represents the probability of observing results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. A small P-value (typically less than 0.05) suggests that the observed results are unlikely to have occurred by chance alone, leading to rejection of the null hypothesis. Conversely, a large P-value (greater than 0.05) indicates that the observed results are plausible under the null hypothesis, and the null hypothesis is not rejected.   


 6)What are Type 1 and Type 2 errors in hypothesis testing?
   - In hypothesis testing, a Type I error (false positive) occurs when you reject a true null hypothesis, and a Type II error (false negative) occurs when you fail to reject a false null hypothesis.


 7)What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
   - In hypothesis testing, a one-tailed test looks for a difference in one specific direction (either greater than or less than), while a two-tailed test looks for a difference in either direction (greater than or less than). This difference in direction affects the placement of the critical region (rejection region) and the interpretation of results.

```python
#Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates z-scores for a dataset and plots them.

    Args:
        data (list or numpy.ndarray): A list or numpy array of numerical data.
    """
    if not isinstance(data, (list, np.ndarray)):
        raise TypeError("Input data must be a list or numpy array.")

    if not all(isinstance(x, (int, float)) for x in data):
         raise ValueError("All elements in the input data must be numerical.")

    data_np = np.array(data)
    mean = np.mean(data_np)
    std_dev = np.std(data_np)

    if std_dev == 0:
        raise ZeroDivisionError("Standard deviation is zero, cannot calculate z-scores.")

    z_scores = (data_np - mean) / std_dev

    plt.figure(figsize=(10, 6))
    plt.scatter(range(len(z_scores)), z_scores)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title('Z-Scores of the Data')
    plt.xlabel('Data Point Index')
    plt.ylabel('Z-Score')
    plt.grid(True)
    plt.show()

# Example usage:
data = [15, 18, 21, 24, 27, 30, 12, 9, 33, 36]
calculate_and_plot_z_scores(data)
```

```python
#Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution(x_start=-4, x_end=4, num_points=1000):
  """
  Calculates and plots the standard normal distribution.

  Args:
    x_start: The starting value for the x-axis range.
    x_end: The ending value for the x-axis range.
    num_points: The number of points to generate within the range.
  """
  x = np.linspace(x_start, x_end, num_points)
  y = norm.pdf(x, 0, 1) # Mean=0, Standard deviation=1

  plt.plot(x, y)
  plt.title('Standard Normal Distribution')
  plt.xlabel('x')
  plt.ylabel('Probability Density')
  plt.grid(True)
  plt.show()

if __name__ == '__main__':
  plot_standard_normal_distribution()
```

```python
#Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Calculates the Z-score for a given data point within a dataset.

    Args:
        data_point (float or int): The data point to calculate the Z-score for.
        data (list or numpy.ndarray): The dataset.

    Returns:
        float: The calculated Z-score.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
  """
  Compares a Z-score to the standard normal distribution.

  Args:
      z_score (float): The Z-score to compare.

  Returns:
      tuple: A tuple containing the probability of observing a value
             as extreme or more extreme than the Z-score (two-tailed test)
             and a string interpretation of the Z-score.
  """
  p_value = 2 * (1 - norm.cdf(abs(z_score))) # Two-tailed p-value
  interpretation = ""

  if abs(z_score) > 3:
    interpretation = "Very unusual (outlier)"
  elif abs(z_score) > 2:
    interpretation = "Unusual"
  elif abs(z_score) > 1:
    interpretation = "Moderately unusual"
  else:
    interpretation = "Typical"

  return p_value, interpretation

if __name__ == "__main__":
    data = [10, 12, 15, 18, 20, 22, 25, 28, 30, 35]
    data_point = 22

    z_score = calculate_z_score(data_point, data)
    p_value, interpretation = compare_to_standard_normal(z_score)

    print(f"Data point: {data_point}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.3f}")
    print(f"Interpretation: {interpretation}")
```

```python
#Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats

# 1. Generate data from a normal distribution
# Set parameters for the normal distribution (mean, standard deviation, sample size)
true_mean = 10  # Population mean
true_std = 2   # Population standard deviation
sample_size = 100  # Number of samples

# Generate random samples from the normal distribution
data = np.random.normal(true_mean, true_std, sample_size)

# 2. Calculate the confidence interval
# Choose the desired confidence level (e.g., 95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# Calculate the z-score corresponding to the confidence level
z_critical = stats.norm.ppf(1 - alpha / 2)

# Calculate the standard error of the mean
sample_mean = np.mean(data)
sample_std = np.std(data)  # Use sample standard deviation if population standard deviation is unknown
std_err = sample_std / np.sqrt(sample_size)

# Calculate the margin of error
margin_of_error = z_critical * std_err

# Calculate the confidence interval bounds
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({confidence_level * 100}%): ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#practical part -2

#Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from scipy import stats

def z_test_sample_mean(sample_data, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test to compare a sample mean to a population mean.

    Args:
        sample_data (list or numpy.ndarray): The sample data.
        pop_mean (float): The known population mean.
        pop_std (float): The known population standard deviation.
        alpha (float, optional): Significance level (default is 0.05).
        alternative (str, optional): Type of test ('two-sided', 'less', 'greater').
                                     Default is 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    if alternative == 'two-sided':
        p_value = 2 * stats.norm.cdf(-np.abs(z_statistic))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    return z_statistic, p_value

# Example usage
sample = [78, 85, 92, 88, 81, 75, 95, 80, 89, 83]
population_mean = 85
population_std_dev = 8
significance_level = 0.05

z_stat, p_val = z_test_sample_mean(sample, population_mean, population_std_dev, alpha=significance_level)

print("Z-statistic:", z_stat)
print("P-value:", p_val)

if p_val < significance_level:
    print("Reject the null hypothesis. There is a significant difference between the sample mean and the population mean.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the sample mean and the population mean.")
```

```python
#Implement a one-sample Z-test using Python to compare the sample mean with the population mean
import numpy as np
from scipy.stats import norm

def one_sample_ztest(sample, popmean, popstd, alternative='two-sided', alpha=0.05):
    """
    Performs a one-sample Z-test.

    Args:
        sample (array-like): The sample data.
        popmean (float): The population mean.
        popstd (float): The population standard deviation.
        alternative (str, optional): The alternative hypothesis.
            Must be one of 'two-sided', 'less', or 'greater'.
            Defaults to 'two-sided'.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - popmean) / (popstd / np.sqrt(n))

    if alternative == 'two-sided':
        p_value = 2 * norm.cdf(-np.abs(z_stat))
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    return z_stat, p_value

# Example usage
sample_data = np.array([155, 160, 145, 150, 165, 140, 152, 158, 148, 162])
population_mean = 150
population_std = 10

z_statistic, p_value = one_sample_ztest(sample_data, population_mean, population_std)

print("Z-statistic:", z_statistic)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
#Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0, mu1, sigma, alpha):
    """
    Visualizes Type I and Type II errors in a one-sided hypothesis test.

    Args:
        mu0 (float):  Mean under the null hypothesis.
        mu1 (float):  Mean under the alternative hypothesis.
        sigma (float): Standard deviation.
        alpha (float): Significance level (Type I error rate).

    Returns:
        None. Displays a plot of the distributions and error regions.
    """

    # 1. Define the distributions
    x = np.linspace(mu0 - 4*sigma, mu0 + 4*sigma, 100)  # Range for x values
    null_dist = norm.pdf(x, mu0, sigma)  # Probability density under null
    alt_dist = norm.pdf(x, mu1, sigma)  # Probability density under alternative

    # 2. Calculate the critical value (rejection region)
    critical_value = norm.ppf(1-alpha, mu0, sigma)  # Inverse CDF for 1-alpha

    # 3. Calculate Type I error probability (false positive)
    type_i_error = norm.cdf(critical_value, mu0, sigma)  # Prob. of rejecting H0 when it's true

    # 4. Calculate Type II error probability (false negative)
    type_ii_error = norm.cdf(critical_value, mu1, sigma)  # Prob. of failing to reject H0 when it's false

    # 5. Plotting the distributions and error regions
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis (H0)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis (H1)", color="orange")
    plt.axvline(x=critical_value, color="red", linestyle="--", label=f"Critical Value (Î±={alpha})")

    # Fill in the error regions
    plt.fill_between(x[x < critical_value], null_dist[x < critical_value], color="red", alpha=0.3, label="Type I Error (False Positive)")
    plt.fill_between(x[x > critical_value], alt_dist[x > critical_value], color="green", alpha=0.3, label="Type II Error (False Negative)")

    plt.title("Visualizing Type I and Type II Errors")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage:
if __name__ == "__main__":
    mu0 = 50  # Mean under null hypothesis
    mu1 = 55  # Mean under alternative hypothesis
    sigma = 5  # Standard deviation
    alpha = 0.05  # Significance level

    visualize_errors(mu0, mu1, sigma, alpha)
```

```python
#Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
def bayes_theorem(prior_a, likelihood_ba, probability_b):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Args:
        prior_a (float): Prior probability of hypothesis A.
        likelihood_ba (float): Likelihood of evidence B given hypothesis A.
        probability_b (float): Probability of evidence B.

    Returns:
        float: Posterior probability of hypothesis A given evidence B.
    """
    posterior_ab = (likelihood_ba * prior_a) / probability_b
    return posterior_ab
```

```python
#Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.
import numpy as np
from scipy.stats import chisquare

# Observed data
observed_values = np.array([25, 30, 20, 25])

# Expected data
expected_values = np.array([20, 35, 25, 20])

# Perform the Chi-squared test
chi2_statistic, p_value = chisquare(f_obs=observed_values, f_exp=expected_values)

print(f"Chi-squared Statistic: {chi2_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```
